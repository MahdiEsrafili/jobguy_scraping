```python
import numpy as np
import pandas as pd
import hazm
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from collections import Counter
from itertools import chain
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split, validation_curve
```

```python
data = pd.read_csv('scrapy_project/jubguy_ir/jubguy_ir/spiders/jobguy_pros_cons.csv', encoding='utf-8')
```

```python
data.tell_friend = data.tell_friend.map(lambda x: 'NNN' if type(x)== float else x)
```

```python
data.head(12)
```

```python
data_review = data[['comment', 'tell_friend']][data.type == 'review']
```

```python
le = LabelEncoder()
```

```python
data_review['tell_fried_labeled'] = le.fit_transform(data_review.tell_friend.values)
```

```python
data_review.head()
```

```python
data_review.comment = data_review.comment.map(lambda x: 'NNN' if type(x)==float else x)
```

```python
data_review['comment_tokenized'] = data_review.comment.map(lambda x: x.split())
```

## create vocabulary of words

```python
full_comments_token =  list(chain(*data_review.comment_tokenized.values))
```

```python
counts = Counter(full_comments_token)
vocab = sorted(counts, key=counts.get, reverse=True)
vocab_to_int = {full_comments_token: ii for ii, full_comments_token in enumerate(vocab, 1)}
```

```python
def token2integer(token):
    '''converts word token array to integer token array '''
    x= [vocab_to_int[t] for t in token]
    return x
```

```python
data_review['integer_token'] = data_review.comment_tokenized.map(token2integer)
```

```python
data_review['summary_token'] = data_review.integer_token.map(lambda x: x[:80] if len(x)>=80 else x+[0]*(80 - len(x)))
```

```python
len(vocab_to_int.keys())
```

```python
data_review.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(data_review.summary_token, data_review.tell_fried_labeled, test_size= 0.3)
```

```python
X_train = [np.array(x) for x in X_train]
X_train = np.array(X_train)
```

```python
X_test = [np.array(x) for x in X_test]
X_test = np.array(X_test)
```

```python
rnd_forrest = RandomForestClassifier(n_estimators=120, max_depth=200)
```

```python
rnd_forrest.fit(X_train, y_train)
```

```python
rnd_forrest.score(X_test, y_test)
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, Dropout, Input, Conv1D, MaxPool1D, Dropout,Flatten
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
```

## CNN

```python
model_cnn=Sequential()
model_cnn.add(Embedding(7813+1,50, input_length=80))
model_cnn.add(Conv1D(64,3,activation='relu' ,padding='valid'))
model_cnn.add(MaxPool1D(2))
model_cnn.add(Conv1D(128,3,activation='relu' ,padding='valid'))
model_cnn.add(MaxPool1D(2))

model_cnn.add(Flatten())
model_cnn.add(Dense(512, activation='sigmoid'))
model_cnn.add(Dense(2, activation='sigmoid'))
model_cnn.compile(loss='sparse_categorical_crossentropy', 
         optimizer='adam', 
         metrics=['accuracy'],learning_rate= 0.001)
num_epochs = 30

es = EarlyStopping(monitor='val_loss', mode='min', verbose=1,patience=5)
model_cnn.fit(X_train,y_train.values,
                        batch_size=128, epochs=num_epochs,
                        validation_data=(X_test,y_test.values),callbacks=[es])
scores =  model_cnn.evaluate(X_test,y_test.values, verbose=0)
print('accuracy:{}'.format(scores[1]))
```
