# Word Cloud of Pros and Cons

```python
from itertools import chain
from collections import Counter
import ast
from PIL import Image
import matplotlib.pylab as plt
import numpy as np 
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import arabic_reshaper
from bidi.algorithm import get_display
```

```python
d= pd.read_csv('scrapy_project/jubguy_ir/jubguy_ir/spiders/jobguy_pros_cons.csv', encoding='utf-8')
```

```python
d
```

```python
pros = d.pros
pros = [ast.literal_eval(x) for x in pros.values]
full_pros = list(chain(*pros))
pros_count = Counter(full_pros)
pros_count = dict(pros_count)
pros_count = {k: v for k, v in sorted(pros_count.items(), key=lambda item: item[1], reverse=True)}
```

```python
cons = d.cons
cons = [ast.literal_eval(x) for x in cons.values]
full_cons = list(chain(*cons))
cons_count = Counter(full_cons)
cons_count = dict(cons_count)
cons_count = {k: v for k, v in sorted(cons_count.items(), key=lambda item: item[1], reverse=True)}
```

```python
mask = np.array(Image.open('mask.png'))
```

```python
cons_mask = np.array(Image.open('cons_mask.png'))
```

```python
new_cons = {}
for k,v in cons_count.items():
    persian_text = arabic_reshaper.reshape(k)
    persian_text = get_display(persian_text)
    new_cons[persian_text] = v
    
```

```python
wordcloud = WordCloud(width=800, height=400,background_color='white',mask = mask, max_words=15, font_path='NotoNaskhArabic-Regular.ttf').generate_from_frequencies(new_cons)
# Display the generated image:
plt.figure( figsize=(20,10) )
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
wordcloud.to_file('cons.png')
```

```python
new_pros = {}
for k,v in pros_count.items():
    persian_text = arabic_reshaper.reshape(k)
    persian_text = get_display(persian_text)
    new_pros[persian_text] = v
```

```python
wordcloud = WordCloud(width=800, height=400,background_color='white', mask= mask, max_words=15, font_path='NotoNaskhArabic-Regular.ttf').generate_from_frequencies(new_pros)
# Display the generated image:
plt.figure( figsize=(20,10) )
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
wordcloud.to_file('pros.png')
```
